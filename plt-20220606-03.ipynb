```python
# matplotlib用途
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
from matplotlib import pyplot
pyplot.rcParams["font.sans-serif"] = ['SimHei']
pyplot.rcParams["axes.unicode_minus"] = False
```

```python
# 直接绘制出精致的图形，并且支持pandas的dataframe
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
x = np.arange(1,10)
y = x**3 - 2*(x**2) +1
```

*1. seaborn直接作为底板*

```python
# 底板调用seaborn配色
plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x, y, linestyle='-.', color='red')
plt.show()
```

*2. 直接在seaborn中设置，底色、放缩的尺寸等，可以设置得更加详细*

```python
sns.set(style='darkgrid', context='notebook', font_scale=1.2)
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x, y, linestyle='-.', color='red')
plt.show()
```

*3. 直接调用seaborn中的接口,要求对象必须是dataframe类型*

```python
df = pd.DataFrame({'cola':x, 'colb':y})
```

```python
sns.lineplot(x='cola',y='colb', data=df,  color='red', linestyle='-')

annot = [(a,b) for a,b in enumerate(df.cola) if a%2==0]
for a,b in annot:
    plt.text(b, df['colb'][a], '%.2f'%df['colb'][a], ha='center', fontsize=13)
    
plt.show()
```
